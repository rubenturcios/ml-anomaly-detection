```python
from pyod.models.iforest import IForest
import pandas as pd
from shap import TreeExplainer
```

```python
DATA_PATH = '../data/event_training_data.csv'

data = pd.read_csv(DATA_PATH)
clf = IForest(n_estimators=1000)
clf.fit(data)

explainer = TreeExplainer(clf, data)
```

```python
data = pd.read_csv('../data/test_df.csv')[['elapsed_time', 'distance', 'revision', 'save_and_exit_count']].iloc[:100]
shap_values = explainer.shap_values(X=data)
```

```python
import shap


shap.initjs()
shap.summary_plot(shap_values, data[['elapsed_time', 'distance', 'revision', 'save_and_exit_count']])
```

```python
sv_df = pd.DataFrame((shap_values.values * -1), columns=['model_impact_elapsed_time_score', 'model_impact_save_and_exit_count_score', 'model_impact_revision_score', 'model_impact_distance_score'])
sv_df.head()
pd.read_json
```
